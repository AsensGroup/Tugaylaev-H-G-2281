```python
dydx = Derivative(y, x) # dydx - имя переменной, обозначающей производную
dydx
```

```python
eq = Eq(dydx - x**2, x)
eq
```

```python
sol = solve(eq, dydx)[0]
sol
```

```python
dsol = integrate(sol, x) + c
dsol
```

```python
x0, y0 = 1, -1
X = dsol.subs(x, x0)
eq = Eq(y0, X)
C0 = solve(eq)[0]
dsol0 = dsol.subs(c, C0)
dsol0
```

```python
p = plot(x ** 2, (x, -1, 1))
```

```python
p = plot(dsol0, (x, -4, 4))
```

```python
p = plot(show=False)
for el in np.arange(-2, 2, 0.25):
  p1 = plot(dsol.subs(c, el), (x, 0,1, 4), show=False)
  p.extend(p1)
p.show()
```
