```python
dydx = Derivative(y, x) # dydx - имя переменной, обозначающей производную
dydx
```

```python
eq = Eq(dydx - x**2, x)
eq
```

```python
sol = solve(eq, dydx)[0]
sol
```

```python
dsol = integrate(sol, x) + c
dsol
```

```python
x0, y0 = 1, -1
X = dsol.subs(x, x0)
eq = Eq(y0, X)
C0 = solve(eq)[0]
dsol0 = dsol.subs(c, C0)
dsol0
```

```python
p = plot(dsol0, (x, -4, 4))
```

```python
from sympy import * # пакет символьных вычислений
import numpy as np # пакет для работы с многомерными массивами
x, y, t, w0, alpha, g = symbols("x y t w0 alpha g")
```

```python
x0 =0
y0 = 0
u0 = w0*cos(alpha)
v0 = w0*sin(alpha)
```

```python
U = 0
V = -g
```

```python
u = integrate(U,t) + u0
v = integrate(V,t) + v0
u, v # выводим на экран найденные интегралы
```

```python
xsol = integrate(u, t) + x0
ysol = integrate(v, t) + y0
xsol, ysol # выводим на экран найденные интегралы
```

```python
t1 = solve(ysol, t)[1]
t1 # вывод на экран
```

```python
x1 = xsol.subs(t, t1)
x1 # вывод на экран
```

```python
T = solve(Eq(xsol, x), t)[0] # выразили t через x
Y = ysol.subs(t, T) # подставили T в y(t) вместо t
Y # вывод формулы на экран
```

```python
ya = Y.subs(g, 8.87).subs(w0, 1).subs(alpha, np.pi/20) # уравнение траектории
x1a = x1.subs(g, 8.87).subs(w0, 1).subs(alpha, np.pi/20) # точка падения
p = plot(ya, (x, 0, x1a))
x1a
```

```python
p = plot(show=False)
for c in np.arange(np.pi/20, np.pi/2, np.pi/20): 
  ya = Y.subs(g, 8.87).subs(w0, 1).subs(alpha, c) # уравнение траектории
  x1a = x1.subs(g, 8.87).subs(w0, 1).subs(alpha, c) # точка падения
  p1 = plot(ya, (x, 0, x1a), show=False)
  p.extend(p1) 
p.show()
```

```python
p = plot(show=False)
for i in range(1, 10):
  for c in np.arange(np.pi/20, np.pi/2, np.pi/20): 
    ya = Y.subs(g, 8.87).subs(w0, i).subs(alpha, c) # уравнение траектории
    x1a = x1.subs(g, 8.87).subs(w0, i).subs(alpha, c) # точка падения
    p1 = plot(ya, (x, 0, x1a), show=False)
    p.extend(p1) 
p.show()
```
